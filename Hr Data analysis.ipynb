<h1>Data Import</h1>

```python
import pandas as pd
from sqlalchemy import create_engine
import matplotlib as plt
```

```python
df=pd.read_csv("D:/Hr Analysis/employee_attrition_dataset_10000.csv",encoding="latin1")
engine=create_engine("postgresql://postgres:20260101@localhost:5432/Hr_Attrition_Analysis")
df.to_sql(
    name="employee_attriton",
    con=engine,
    if_exists="replace",
    index=False
)
print("Data import in SQL Successfully!")
```
